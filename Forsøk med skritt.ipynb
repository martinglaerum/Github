```python
from pylab import*

data = loadtxt("data.csv", skiprows = 1, delimiter = ",")

tid = data[:,0]
a = data[:,3]

skritt = 0

plot(tid,a)
show()

for i in range(0,3863):
    if a[i] > 3 and a[i-1] < 3:
        skritt = skritt + 1
        
print(skritt)
```
