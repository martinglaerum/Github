```python
from pylab import *

t = 48
a = 4
Nmaks = 157759
N0 = 3
S = Nmaks-N0
Tid = []
Nliste = []
Sliste = []

for i in range(t):
    a = 4/Nmaks
    N = N0 + a*N0*S
    N0 = N
    S = S-a*N0*S
    Tid.append(i)
    Nliste.append(N)
    Sliste.append(S)
    
plot(Tid,Nliste,label="Smittede")
plot(Tid,Sliste,label="Mottakelige")
legend()
show()
```

```python
from pylab import *

t = 48
R = 0
a = 4
b = 3.5
Nmaks = 157759
N0 = 3
S = Nmaks-N0
Tid = []
Nliste = []
Recovered = []
Sliste = []

for i in range(t):
    a = 4/Nmaks
    N = N0 + a*N0*S-b*N0
    N0 = N
    S = S-a*N0*S
    R = R+b*N0
    Tid.append(i)
    Nliste.append(N)
    Recovered.append(R)
    Sliste.append(S)
    
plot(Tid,Nliste,label="Smittede")
plot(Tid,Recovered,label="Restituerte")
plot(Tid,Sliste,label="Mottakelige")
legend()
show()
```
